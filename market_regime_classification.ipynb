```python
df = pd.DataFrame()
df['Time'] = data['Time']
```

```python
# feature engineering
# liquidity & depth features
for level in range(1,21):
    bid_col = f"BidPriceL{level}"
    ask_col = f"AskPriceL{level}"
    spread_col = f"Bid/Ask_SpreadL{level}"
    df[spread_col] = data[ask_col] - data[bid_col]
    
```

```python
# orderbook imbalance at each level: imbalance,microprice

for level in range(1,21):
    bid_col = f"BidQtyL{level}"
    ask_col = f"AskQtyL{level}"
    imbalance = f"imbalance_lvl{level}"
    df[imbalance] = (data[bid_col] - data[ask_col])/(data[bid_col] + data[ask_col])
```

```python
for level in range(1,21):
    bid_price = f"BidPriceL{level}"
    ask_price = f"AskPriceL{level}"
    bid_Qty = f"BidQtyL{level}"
    ask_Qty = f"AskQtyL{level}"
    microprice_col = f"microprice_lvl{level}"
    df[microprice_col] = (data[bid_price]*data[ask_Qty] + data[ask_price]*data[bid_Qty])/(data[bid_Qty] + data[ask_Qty])
```

```python
#cumulative depth
ask_sum = 0
bid_qty_columns = [f"BidQtyL{level}" for level in range(1,21)]
ask_qty_columns = [f"AskQtyL{level}" for level in range(1,21)]
df["cum_bid_qty"] = data[bid_qty_columns].sum(axis=1)   
df["cum_ask_qty"] = data[ask_qty_columns].sum(axis=1)  
```

```python
# volatility & price action Rolling mid-price return, price volatlity
for level in range(1,21):
    bid_col = f"BidPriceL{level}"
    ask_col = f"AskPriceL{level}"
    mid_price = f"mid_priceL{level}"
    rolling_price = f"rolling_priceL{level}"
    df[mid_price] = (data[bid_col] + data[ask_col])/2
    

```

```python
data.Time.dtype
```

```python
# as the data is for 25-03-14 and also +0530 IST we can remove that part from the time column
```

```python
df['Time'] = df['Time'].str.slice(10,29)
```

```python
df.head(-1)
```

```python
for level in range(1,21):
    bid_col = f"BidPriceL{level}"
    ask_col = f"AskPriceL{level}"
    mid_price = f"mid_priceL{level}"
    rolling_returns = f"rolling_returnsL{level}"
    df[rolling_returns] = np.log(df[mid_price] / df[mid_price].shift(1))
```

```python
df.head()
```

```python
price_cols = [f'AskPriceL{i}' for i in range(1, 21)]
qty_cols = [f'AskQtyL{i}' for i in range(1, 21)]

df['VWAP_numerator'] = sum(data[p] * data[q] for p, q in zip(price_cols, qty_cols))
df['VWAP_denominator'] = sum(data[q] for q in qty_cols)

df['VWAP'] = df['VWAP_numerator'] / df['VWAP_denominator']
```

```python
df['VWAP_Change'] = df['VWAP'] - df['VWAP'].shift(1)
```

```python
df1 = df.dropna()
```

```python
df1.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression

levels = np.arange(1, 21).reshape(-1, 1)
bid_qty_cols = [f'BidQtyL{i}' for i in range(1, 21)]

def compute_slope(row):
    qtys = row[bid_qty_cols].values.reshape(-1, 1)
    reg = LinearRegression().fit(levels, qtys)
    return reg.coef_[0][0]  # slope

df1['BidSlope'] = data.apply(compute_slope, axis=1)
```

```python
df1['BidSlope'].describe()
```

```python
# even distribution
```

```python
# Slope Value | Meaning
# -ve & large | Heavy liquidity near top
# ≈ 0 | Even distribution
# +ve (rare) | More size deeper (illiquid)
```

```python
from sklearn.linear_model import LinearRegression

levels = np.arange(1, 21).reshape(-1, 1)
ask_qty_cols = [f'AskQtyL{i}' for i in range(1, 21)]

def compute_slope(row):
    qtys = row[ask_qty_cols].values.reshape(-1, 1)
    reg = LinearRegression().fit(levels, qtys)
    return reg.coef_[0][0]  # slope

df1['AskSlope'] = data.apply(compute_slope, axis=1)
```

```python
df1['AskSlope'].describe()
```

```python
# even distribution
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df2 = df1.drop(['Time'],axis=1)
```

```python
df2.columns
```

```python
sc = StandardScaler()
df3 = sc.fit_transform(df2)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca_data = pca.fit_transform(df3)

pca_df = pd.DataFrame(data=pca_data, columns=[f'PC{i+1}' for i in range(pca_data.shape[1])])


print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)

pca_df.head()
```

```python
from sklearn.cluster import KMeans


kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(pca_df)


pca_df['Cluster'] = clusters
```
